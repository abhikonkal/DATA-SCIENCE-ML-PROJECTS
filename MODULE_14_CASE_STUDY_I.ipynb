```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('bio-degradabale-data.csv',sep=';')
df.head()
```

```python
df.shape
```

```python
df.columns=range(42)
```

```python
df.head()
```

```python
df[41]=df[41].map({'RB':1,'NRB':0})
```

```python
df.head()
```

```python
X = df.iloc[:,0:41]
y = df[41]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=85)
```

```python
clf=DecisionTreeClassifier(random_state=42)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

ADA boosting

```python
boost_clf=AdaBoostClassifier(base_estimator=clf,learning_rate=.2,n_estimators=50)
```

```python
boost_clf.fit(X_train,y_train)
y_pred_boost=boost_clf.predict(X_test)
accuracy_score(y_test,y_pred_boost)
```
