```python
data=np.genfromtxt('SalaryGender.csv',delimiter=',',skip_header=1)

#as any sepcificationa was not given i am assuming men -->1 and women-->zero
a=len(data)
phdandgender=data[:a,::-2]
#genderandphd.reshape(2,a)
print(phdandgender[1][1])

mancount=0
womencount=0
phdman=0
phdwomen=0

for i in range(0,a):
    if phdandgender[i][1]==1:
        mancount+=1
        if phdandgender[i][0]==1:
            phdman+=1
    else:
        womencount+=1
        if phdandgender[i][0]==1:
            phdwomen+=1
print(f'no of men : {mancount} No of men with phd: {phdman}')
print(f'no of women : {a-1-mancount} No of women with phd: {phdwomen}')


```

##reading and decribing data

```python
df=pd.read_csv('middle_tn_schools.csv')
df_grouped_data = df.groupby("school_rating").describe()
df_grouped_data
```

Correlation analysis

```python
cormat=df_grouped_data.corr()
cormat.describe()
```

scatter plot AND SNS HEATMAP

```python
plt.figure(figsize=(10,10))
sns.heatmap(cormat,cmap='YlGnBu',linewidths=0.5)
plt.scatter(x=df["school_rating"],y=df["reduced_lunch"])
plt.grid()
plt.xlabel("Rating")
plt.ylabel("Reduced lunch")
```
