```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Loading Data into Flies

```python
dfp=pd.read_csv('PhysicsScoreTerm1.csv')

dfm=pd.read_csv('MathScoreTerm1.csv')

dfds=pd.read_csv('DSScoreTerm1.csv')

display(dfp,dfm,dfds)

name=dfp['Name']
ethnicity=dfp['Ethinicity']
```

Removing name and Ethinicity Coloumn

```python
dfp=dfp.drop(labels=['Name','Ethinicity'],axis=1)
dfm=dfm.drop(labels=['Name','Ethinicity'],axis=1)
dfds=dfds.drop(labels=['Name','Ethinicity'],axis=1)
```

```python
display(dfp,dfm,dfds)
```

Filling Missing Data with Zeroes

```python
dfp['Score'].replace(np.nan,0)
dfm['Score'].replace(np.nan,0)
dfds['Score'].replace(np.nan,0)
print(dfp,dfm,dfds)
```

Merging Data

```python
df_temp=pd.merge(dfp,dfm,on=['ID','Sex','Age'])

#df_temp.rename(columns={'Score_x':'Phy_marks','Score_y':'Math_marks','Subject_x':'Physics','Subject_y':'Maths'})

df_final=pd.merge(df_temp,dfds,on=['ID','Sex','Age'])

df_final.rename(columns={'Score_x':'Phy_marks','Score_y':'Math_marks','Subject_x':'Physics','Subject_y':'Maths'},inplace=True)

df_final
```

Changing M/F to 1/2

```python
df_final.loc[df_final['Sex']=='M','Sex']=1 
df_final.loc[df_final['Sex']=='F','Sex']=2 
df_final
```

result saved

```python
result_file=df_final.to_csv('Result_m4_case_study-ii.csv')
```

Enhancements

```python

df_final.insert(0,'Name',name)
df_final.insert(1,'Ethinicity',ethnicity)
df_final
```

Ethinicity is added and now converting it into numericals

```python
setval=set(df_final['Ethinicity'])
setval
```

```python

df_final['Ethinicity'].replace({'White American':100,'European American':101,'Hispanic':103,'African American':104,},inplace=True)
df_final
```

Replacing null with score mean

```python
df_final['Math_marks'].fillna(df_final['Math_marks'].mean(),inplace=True)
df_final['Phy_marks'].fillna(df_final['Phy_marks'].mean(),inplace=True)
df_final['Score'].fillna(df_final['Score'].mean(),inplace=True)
df_final
```
