```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
```

```python
df_main=pd.read_csv('loan_borowwer_data.csv')
df=df_main
df.head()
```

```python
df.shape
```

```python
df[df['credit.policy']==0]
```

```python
df.drop(df.index[df['credit.policy']==0])
```

```python
X=df.drop(['not.fully.paid','credit.policy','purpose'],axis=1)
y=df['not.fully.paid']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=45,test_size=0.3,stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred=rfc.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
