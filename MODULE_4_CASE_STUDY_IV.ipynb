```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('BigMartSalesData.csv')
df.head()
```

Q no 1

```python
df_2011=df[df['Year']==2011]
```

```python
data=df_2011.groupby(df_2011['Month']).agg({'Amount':np.sum})
print(data)
low=data['Amount'].idxmin()
print(f'{low} month  has lowest stocks.this refelcts in the graph.')
data.plot(kind='line',color='g',marker='o')
plt.xlabel('Months of 2011')
plt.ylabel('Amount')
plt.title('STOCK IN 2011 Monthwise')
```

Q NO 2

BAR plot

```python

data=df_2011.groupby(df_2011['Month'],as_index=False).agg({'Amount':np.sum})

plt.barh(data['Month'],data['Amount'])

plt.xlabel('Months of 2011')
plt.ylabel('Amount')
plt.title('STOCK IN 2011 Monthwise')
```

simple plain plot is not that good looking 
but line plot has trend show 
and bar plot looks clean and shows in a well distributed manner.

Q NO 3

country wise pie plot 

```python
data=df_2011.groupby(df_2011['Country']).agg({'Amount':np.sum})
print(data.head())
high=data['Amount'].idxmax()
print(f'{high} has highest conntributions stocks.this refelcts in the graph.')

#shadow true are added
plt.pie(data['Amount'],labels=data['Amount'].index,shadow=True,startangle=90,radius=3,autopct='%0.2f%%')

plt.title('STOCK IN 2011 countrywise')
```
