```python
df2=df1
df2['total_benefitted']=df1.iloc[:,2:6].sum(axis=1)
df2.head(10)
```

b

Q NO 3

a

```python
x=df2['STATE/UT']
y=df2['total_benefitted']
dft=df2.groupby('STATE/UT').agg({'total_benefitted':np.sum})
mcoutry=dft['total_benefitted'].idxmax()
plt.bar(x,y)
plt.title("STATE/UT vs Total_inmates_benefitted")
plt.xlabel("STATE/UT")
plt.ylabel("Total_inmates_benefitted")
plt.show()
print(f'Country with maximum beneficiries is : {mcoutry}.')
```

b

```python
plt.pie(y,labels=x,autopct='%0.2f%%')
plt.title('Pie-Chart')
```
